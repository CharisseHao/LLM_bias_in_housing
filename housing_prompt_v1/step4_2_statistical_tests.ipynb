Outputs: no files, only results in notebook

```python
import sys
sys.path.append('../src/')
from stat_tests import assumptions_and_tests, detailed_dunns_test_bivariate, detailed_dunns_test

import pandas as pd
import seaborn as sns
import datetime


import warnings
warnings.filterwarnings("ignore")

sns.set_theme(context='notebook', style='whitegrid')
pd.set_option("display.max_rows", 100)
```

```python
start_run = datetime.datetime.now()
```

```python
df = pd.read_csv("processed_data/housing_prompt_v1_allmodels.csv.zip", compression='zip')
df_valid = df[df['query_response'].notna()]
df_valid.head()
```

## Single Variable

```python
variables = ['gender', 'race', 'occupation', 'living_status']
models = df_valid['model'].unique()
assumptions_and_tests(df_valid, 'query_response', models, variables)
```

## Multiple Variables

```python
# occupation x gender
var1 = 'occupation'
var2 = 'gender'

for m in models:
    print(f"\033[1mmodel: {m}\033[0m")
    model_df = df_valid[df_valid['model'].str.contains(m)]
    display(detailed_dunns_test_bivariate(df_valid, 'query_response', var1, var2, total_comparisons=1000))
```
